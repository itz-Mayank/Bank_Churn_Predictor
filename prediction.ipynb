```python
import tensorflow as tf
import pandas as pd
import numpy as np
import pickle
from tensorflow.keras.models import load_model
```

```python
# import scaler, OHE, model pickel files

model = load_model('model.h5')

with open('OHE_Geo.pkl','rb') as file:
    geo_encoder = pickle.load(file)
    
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('Scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000,
}
```

```python
geo_encode = geo_encoder.transform([[input_data['Geography']]]).toarray()
geo_encode_df = pd.DataFrame(geo_encode.astype(int),columns=geo_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encode_df
```

```python
input_data = pd.DataFrame([input_data])
```

```python
input_data
```

```python
input_data['Gender'] = label_encoder_gender.transform(input_data['Gender'])
input_data
```

```python
input_data = pd.concat([input_data.drop('Geography',axis = 1),geo_encode_df],axis = 1)
```

```python
input_data
```

```python
# Scaling the input
scaled_df = scaler.transform(input_data)
```

```python
scaled_df
```

```python
prediction = model.predict(scaled_df)
prediction
```
